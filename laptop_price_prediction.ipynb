# Laptop Price Prediction

### Importing required libraries

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk
%matplotlib inline
```

### Loading dataset

```python
data = pd.read_csv('laptopPrice.csv')
```

### Data exploration

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

### Relationship between brand and price

```python
data['brand'].unique()
```

```python
sns.scatterplot(x=data['brand'],y=data['Price'])
```

### Categorical data

```python
cat = data.columns[data.dtypes=='object']
```

```python
cat
```

### Converting into numerical data

```python
data = pd.get_dummies(data,columns=cat,dtype=int)
```

```python
data
```

### Splitting feature and target data

```python
X = data.drop('Price',axis=1)
y = data['Price']
```

```python
X
```

```python
y
```

### Rrain test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

### Regression line

```python
plt.scatter(y_test,y_pred,color='orange')
plt.plot(y_test,y_test,color='blue',lw=1.5)
```

### Checking the score

```python
lr.score(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```
